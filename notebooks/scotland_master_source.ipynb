```python
import pandas as pd
```

```python
# Imports CSVs
election_results_df = pd.read_csv('../processed_data/scotland_election_results_dataframe.csv')
scot_polls = pd.read_csv('../processed_data/scot_polls.csv')
```

```python
# Renames column names
election_results_df.rename(columns={'Year': 'election_year'}, inplace=True)
```

```python
scot_polls['election_year'] = scot_polls['next_elec_date'].str[:4].astype(int)
```

```python
# Merges dataframes
scot_polls_actuals = scot_polls.merge(election_results_df, on='election_year', how='left')
```

```python
rename_columns = {
    'BRX': 'BRX_FC',
    'CON': 'CON_FC',
    'GRE': 'GRE_FC',
    'LAB': 'LAB_FC',
    'LIB': 'LIB_FC',
    'NAT': 'NAT_FC',
    'OTH': 'OTH_FC',
    'PLC': 'PLC_FC',
    'SNP': 'SNP_FC',
    'UKI': 'UKI_FC',
    'BRX_ACTUAL_PERCENTAGE': 'BRX_ACT',
    'CON_ACTUAL_PERCENTAGE': 'CON_ACT',
    'GRE_ACTUAL_PERCENTAGE': 'GRE_ACT',
    'LIB_ACTUAL_PERCENTAGE': 'LIB_ACT',
    'LABOUR_ACTUAL_PERCENTAGE': 'LAB_ACT',
    'NAT_ACTUAL_PERCENTAGE': 'NAT_ACT',
    'PLC_ACTUAL_PERCENTAGE': 'PLC_ACT',
    'SNP_ACTUAL_PERCENTAGE': 'SNP_ACT',
    'UKI_ACTUAL_PERCENTAGE': 'UKI_ACT'
}
```

```python
scot_polls_actuals.rename(columns=rename_columns, inplace=True)
```

```python
scot_polls_actuals.drop(columns=['Country'], inplace=True)
```

```python
combined_df = scot_polls_actuals
```

```python
combined_df.rename(columns={'pollster_rating': 'rating'}, inplace=True)
```

```python
combined_df.rename(columns={'next_election_date': 'next_elec_date'}, inplace=True)
```

```python
combined_df.rename(columns={'days_until_next_election': 'days_to_elec'}, inplace=True)
```

```python
combined_df.drop(columns='Geography', inplace=True)  # Drop column
```

```python
combined_df.drop(columns='election_year', inplace=True)  # Drop column
```

```python
combined_df.to_csv('../processed_data/scotland_polling_results_cleaned.csv', index=True)
```
