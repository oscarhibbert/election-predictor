```python
import pandas as pd
```

```python
# Load constituency naming files
constituencies_2010 = pd.read_csv('../raw_data/PCON_DEC_2021_UK_NC-Table 1.csv')
constituencies_2024 = pd.read_csv('../raw_data/Westminster_Parliamentary_Constituencies_(Future)_Names_and_Codes_in_the_United_Kingdom_v2.csv')
mapping_file = pd.read_csv('../raw_data/Westminster_PCON_(2010)_to_future_Westminster_PCON_(2024)_Lookup_in_the_UK_(V2).csv')
```

```python
mapping_file.head()
```

```python
# Drop and rename columns
constituencies_2010.drop(columns=['Unnamed: 2',	'Unnamed: 3', 'Unnamed: 4'],inplace=True)
# constituencies_2010.rename(columns={'PCON21CD':'ID','PCON21NM':'Old Name'},inplace=True)

constituencies_2024.drop(columns=['PCON24NMW', 'ObjectId'],inplace=True)
# constituencies_2024.rename(columns={'PCON24CD':'ID','PCON24NM':'New Name'},inplace=True)
```

```python
# Join 2 files
constituency_mapping = pd.merge(constituencies_2024,mapping_file,how='left',left_on='PCON24CD',right_on='PCON24CD')
```
